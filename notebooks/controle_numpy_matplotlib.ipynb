# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
import numpy as np
import matplotlib.pyplot as plt
#importation du dataset
# changement du type des données au <20
data = np.genfromtxt('..\data\Iris.csv', delimiter=',', skip_header=1, dtype='<U20')

print(f"shape du dataset : {data.shape}")
```

### E26. Line plot superposé
Cumsum des largeurs des pétales par espèce.


```python
plt.figure(figsize=(10, 6))

# Calcul des cumsum pour chaque espèce (Colonne 3 = Petal Width)
cumsum_setosa = np.cumsum(features[:50, 3])
cumsum_versicolor = np.cumsum(features[50:100, 3])
cumsum_virginica = np.cumsum(features[100:, 3])

# Tracé des 3 courbes
# On ne met pas d'axe X specifique, donc ce sera l'index (0 à 49)
plt.plot(cumsum_setosa, label='Setosa', color='red', linestyle='-')
plt.plot(cumsum_versicolor, label='Versicolor', color='green', linestyle='--')
plt.plot(cumsum_virginica, label='Virginica', color='blue', linestyle='-.')

# Ajout des labels et titre
plt.title("Vitesse d'accumulation de la matière (Largeur Pétale)")
plt.xlabel("Nombre d'échantillons")
plt.ylabel("Cumul (cm)")
plt.legend()

# Affichage du graphique
plt.show()

# INTERPRETATION DU RESULTAT (Commentaire analytique)
# - La courbe Rouge (Setosa) est tres plate : cela signifie que chaque fleur ajoute tres peu
#   a la somme (petales tres fins, ~0.2 cm).
# - La courbe Bleue (Virginica) a une pente tres raide : cela signifie que chaque fleur
#   ajoute beaucoup a la somme (petales larges, >2.0 cm).
# - Conclusion : La pente de la courbe cumsum nous indique la "taille moyenne" de l'espece.
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# 1. Calcul manuel des variances
# On calcule la variance totale pour chaque groupe
var_setosa = np.sum(np.var(features[:50], axis=0))
var_versicolor = np.sum(np.var(features[50:100], axis=0))
var_virginica = np.sum(np.var(features[100:], axis=0))

print("Variance Setosa :", var_setosa)
print("Variance Versicolor :", var_versicolor)
print("Variance Virginica :", var_virginica)
print("L'espece avec la plus grande variance est : Virginica")

# 2. Visualisation cote a cote (3 sous-graphiques)
plt.figure(figsize=(18, 5))

# --- Graphique 1 : Setosa (0 a 50) ---
plt.subplot(1, 3, 1)
# On trace les 4 histogrammes superposes pour Setosa
plt.hist(features[:50, 0], alpha=0.5, label='Sepal Len')
plt.hist(features[:50, 1], alpha=0.5, label='Sepal Wid')
plt.hist(features[:50, 2], alpha=0.5, label='Petal Len')
plt.hist(features[:50, 3], alpha=0.5, label='Petal Wid')
plt.title("Setosa (Variance Faible)")
plt.xlabel("Taille (cm)")
plt.ylabel("Frequence")
plt.legend()

# --- Graphique 2 : Versicolor (50 a 100) ---
plt.subplot(1, 3, 2)
# On trace les 4 histogrammes superposes pour Versicolor
plt.hist(features[50:100, 0], alpha=0.5, label='Sepal Len')
plt.hist(features[50:100, 1], alpha=0.5, label='Sepal Wid')
plt.hist(features[50:100, 2], alpha=0.5, label='Petal Len')
plt.hist(features[50:100, 3], alpha=0.5, label='Petal Wid')
plt.title("Versicolor (Variance Moyenne)")
plt.xlabel("Taille (cm)")

# --- Graphique 3 : Virginica (100 a 150) ---
plt.subplot(1, 3, 3)
# On trace les 4 histogrammes superposes pour Virginica
plt.hist(features[100:, 0], alpha=0.5, label='Sepal Len')
plt.hist(features[100:, 1], alpha=0.5, label='Sepal Wid')
plt.hist(features[100:, 2], alpha=0.5, label='Petal Len')
plt.hist(features[100:, 3], alpha=0.5, label='Petal Wid')
plt.title("Virginica (Variance Max)")
plt.xlabel("Taille (cm)")

plt.show()
```
